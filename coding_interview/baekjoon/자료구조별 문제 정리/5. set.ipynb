```python
import sys
sys.setrecursionlimit(10**9)
input = sys.stdin.readline
n,m = map(int,input().split())
def find_parent(x):
    if sets[x] != x:
        sets[x] = find_parent(sets[x])
    return sets[x]

sets = [i for i in range(n+1)]
for _ in range(m):
    oper, a,b = map(int, input().split())
    a,b = find_parent(a),find_parent(b)
    if not oper:
        if sets[a] > sets[b] : sets[a] = sets[b]
        else: sets[b] = sets[a]
    else:
        if a == b:
            print('YES')
        else:
            print('NO')
```

```python
# 20040번 : 사이클 게임
# 
from sys import stdin
sys.setrecursionlimit(10**9)
input = sys.stdin.readline

def find_parent(x):
    if sets[x] != x:
        sets[x] = find_parent(sets[x])
    return sets[x]

n,m = map(int,input().split())
sets = [i for i in range(n)]
ans = 0

for i in range(1,m+1):
    a,b = map(int,input().split())
    a_root = find_parent(a)
    b_root = find_parent(b)

    if a_root < b_root:
        sets[b_root] = a_root
    elif a_root > b_root:
        sets[a_root] = b_root
    else:
        ans = i
        break
print(ans)
```
