```python
# 1774번 : 별자리 만들기
import sys
sys.stdin = open('input.txt')
# input 정의가 sys.stdin밑에 있어야 텍스트를 읽을 수 있음
input = sys.stdin.readline

def find_parent(x):
    if f_union[x] != x:
        f_union[x] = find_parent(f_union[x])
    return f_union[x]

n = int(input())
stars = [list(map(float,input().split())) for _ in range(n)]
f_union = [i for i in range(n)]
edges = []
# 각 edges의 길이 구하기
for i in range(len(stars)):
    for l in range(i+1,len(stars)):
        x1,y1 = stars[i]
        x2,y2 = stars[l]
        edges.append((i,l,((x1 - x2)**2 + (y1-y2)**2)**(1/2)))

edges.sort(key=lambda x : x[2])

ans = 0
for edge in edges:
    n1,n2,c = edge
    n1 = find_parent(n1)
    n2 = find_parent(n2)

    if n1 == n2: continue
    elif n1 > n2:
        f_union[n2] = n1
        ans += c
    else:
        f_union[n1] = n2
        ans += c

print('{:.2f}'.format(ans))
```

```python
# 1753번 : 최단경로
import sys
from heapq import heappush,heappop
input = sys.stdin.readline
INF = sys.maxsize

n,e = map(int,input().split())
st = int(input())
edges = [ [] for _ in range(n + 1)]
distance = [ INF for _ in range(n + 1)]
distance[st] = 0

for _ in range(e):
    u,v,w = map(int,input().split())
    edges[u].append((w,v))

h = []
heappush(h,(0,st))

while h:
    # w는 st부터 now까지 알려진 최단 거리 
    w, now = heappop(h)

    # 다익스트라 밑에 조건이 이해가 안감
    if distance[now] < w:
        continue
    else:
        for edge in edges[now]:
            # for문으로 2차원 리스트에서 1차원 리스트를 하나씩 불러올 때 
            if distance[edge[1]] > w + edge[0]:
                distance[edge[1]] = w + edge[0]
                heappush(h,[distance[edge[1]],edge[1]])

    

for num in distance[1:]:
    if num == INF:
        print('INF')
    else: print(num)
```

```python
# 1504번 : 특정한 최단 경로 (경유지가 있는 경우 다익스트라)
# 거쳐야 하는 특정 경로를 대상으로 다익스트라 알고리즘을 여러번 사용
import sys
from heapq import heappush,heappop
input = sys.stdin.readline
INF = sys.maxsize


def shortcut(st,end,n):
    dis = [INF for _ in range(n+1)]
    dis[st] = 0
    heap = [[0,st]]
    while heap:
        w_now, now = heappop(heap)
        if dis[now] < w_now:
            continue
        else:
            for w_next, next in edges[now]:
                if dis[next] > w_now + w_next:
                    dis[next] = w_now + w_next
                    heappush(heap,[dis[next], next])
    return dis[end]


# 1번 정점에서 n번 정점으로 최단 거리로 이동
n,m = map(int,input().split())
edges = [[] for _ in range(n+1)]
st = 1
end = n
for _ in range(m):
    e1,e2,w = map(int,input().split())
    edges[e1].append((w,e2))
    edges[e2].append((w,e1))

v1,v2 = map(int,input().split())

st_v1 = shortcut(st,v1,n)
st_v2 = shortcut(st,v2,n)
v1_v2 = shortcut(v1,v2,n)
v1_end = shortcut(v1,end,n)
v2_end = shortcut(v2,end,n)

if min(st_v1 + v1_v2 + v2_end,st_v2 + v1_v2 + v1_end) >= INF : print(-1)
else : print(min(st_v1 + v1_v2 + v2_end,st_v2 + v1_v2 + v1_end))
```

```python
# 1976번 : 
# 유니온 파인드 구현 
sys.setrecursionlimit(10**9)
sys.stdin = open('./input.txt')
input = sys.stdin.readline

def find_parent(r):
    if city[r] != r:
        city[r] = find_parent(city[r])
    return city[r]

n = int(input())
city = [ i for i in range(n+1)]
m = int(input())
maps = [list(map(int,input().split())) for _ in range(n)]
routes = list(map(int,input().split()))

for r in range(n):
    for c in range(r+1,n):
        if maps[r][c]:
            row,col = find_parent(r+1),find_parent(c+1)
            if city[row] > city[col] : city[row] = city[col]
            else: city[col] = city[row]

ans = find_parent(routes[0])
for route in routes[1:]:
    if ans != find_parent(route):
        print('NO')
        exit()
print('YES')
```
