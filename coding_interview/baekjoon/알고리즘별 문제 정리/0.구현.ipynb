```python
# 14502번 : 연구소 
# 3개의 벽을 모두 놓을 수 있는 곳을 모두 계산하여 최대값을 구한다. 
import sys
import copy
from collections import deque
input = sys.stdin.readline


def bfs(virus, walls):
    b = copy.deepcopy(board)
    # 벽 세우기
    for r,c in walls:
        b[r][c] = 1
    q = deque(virus)
    # 바이러스 퍼트리기
    move_r = [1,-1,0,0]
    move_c = [0,0,1,-1]
    visited = [[ 0 for _ in range(m)] for _ in range(m)]
    while q:
        now_r, now_c = q.popleft()
        if visited[now_r][now_c] == 0:
            visited[now_r][now_c] = 1
            for i in range(4):
                next_r = now_r + move_r[i]
                next_c = now_c + move_c[i]
                if 0 <= next_r < n and 0 <= next_c < m:
                    if b[next_r][next_c] == 0 and not visited[next_r][next_c]:
                        b[next_r][next_c] = 2
                        q.append((next_r,next_c))

    # 빈 공간 세기
    ans = 0
    for a in b:
        ans += a.count(0)

    return ans

board = []
n,m = map(int,input().split())

for _ in range(n):
    board.append(list(map(int,input().split())))

virus = []
empty = []
for row in range(n):
    for col in range(m):
        if board[row][col] == 2:
            virus.append((row,col))
        elif board[row][col] == 0:
            empty.append((row,col))

combis = list(combinations(empty,3))
ans = -1
for combi in combis:
    a = bfs(virus,combi)
    if a > ans:
        ans = a


print(ans)
```
