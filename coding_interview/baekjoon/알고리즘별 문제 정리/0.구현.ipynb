```python
# 14
import sys
input = sys.stdin.readline

def move(option):
    dice_tmp = dice.copy()
    # move_east
    if option == 1:
        # top -> right -> bottom -> left -> top
        dice[right] = dice_tmp[top]
        dice[bottom] = dice_tmp[right]
        dice[left] = dice_tmp[bottom]
        dice[top] = dice_tmp[left]
    # move_west
    elif option == 2:
        # top -> left -> bottom -> right -> top
        dice[left] = dice_tmp[top]
        dice[bottom] = dice_tmp[left]
        dice[right] = dice_tmp[bottom]
        dice[top] = dice_tmp[right]

    # move_north
    elif option == 3:
        # top -> up -> bottom -> down -> top
        dice[up] = dice_tmp[top]
        dice[bottom] = dice_tmp[up]
        dice[down] = dice_tmp[bottom]
        dice[top] = dice_tmp[down]

    #move_south
    elif option == 4:
        # top -> down -> bottom -> up -> top
        dice[down] = dice_tmp[top]
        dice[bottom] = dice_tmp[down]
        dice[up] = dice_tmp[bottom]
        dice[top] = dice_tmp[up]

def is_onboard(r,c,option):
    moves = [(0,0),(0,1),(0,-1),(-1,0),(1,0)]
    next_r = r + moves[option][0]
    next_c = c + moves[option][1]
    if next_r >= 0 and next_r < n and next_c >= 0 and next_c < m:
        return True
    return False

def copy_num(r,c):
    if board[r][c] == 0:
        board[r][c] = dice[1]
    else:
        dice[1] = board[r][c]
        board[r][c] = 0


n,m,r,c,k = map(int,input().split())
board = []
for _ in range(n):
    board.append(list(map(int,input().split())))

top, bottom, left, right, up ,down = 0, 1, 2, 3, 4, 5
dice = [0, 0, 0, 0, 0, 0]
cmds = list(map(int,input().split()))
moves = [(0, 0), (0, 1), (0, -1), (-1, 0), (1, 0)]

for cmd in cmds:
    if is_onboard(r,c,cmd):
        r = r + moves[cmd][0]
        c = c + moves[cmd][1]

        move(cmd)
        copy_num(r,c)
        print(dice[0])


```

```python
# 14502번 : 연구소 
# 3개의 벽을 모두 놓을 수 있는 곳을 모두 계산하여 최대값을 구한다. 
import sys
import copy
from collections import deque
input = sys.stdin.readline


def bfs(virus, walls):
    b = copy.deepcopy(board)
    # 벽 세우기
    for r,c in walls:
        b[r][c] = 1
    q = deque(virus)
    # 바이러스 퍼트리기
    move_r = [1,-1,0,0]
    move_c = [0,0,1,-1]
    visited = [[ 0 for _ in range(m)] for _ in range(m)]
    while q:
        now_r, now_c = q.popleft()
        if visited[now_r][now_c] == 0:
            visited[now_r][now_c] = 1
            for i in range(4):
                next_r = now_r + move_r[i]
                next_c = now_c + move_c[i]
                if 0 <= next_r < n and 0 <= next_c < m:
                    if b[next_r][next_c] == 0 and not visited[next_r][next_c]:
                        b[next_r][next_c] = 2
                        q.append((next_r,next_c))

    # 빈 공간 세기
    ans = 0
    for a in b:
        ans += a.count(0)

    return ans

board = []
n,m = map(int,input().split())

for _ in range(n):
    board.append(list(map(int,input().split())))

virus = []
empty = []
for row in range(n):
    for col in range(m):
        if board[row][col] == 2:
            virus.append((row,col))
        elif board[row][col] == 0:
            empty.append((row,col))

combis = list(combinations(empty,3))
ans = -1
for combi in combis:
    a = bfs(virus,combi)
    if a > ans:
        ans = a


print(ans)
```

```python
# 18405번 : 경쟁적 전염
# 여러개의 점에서 bfs를 동시에 진행하는 문제 
import sys
input = sys.stdin.readline

n,k = map(int,input().split())
board = [list(map(int,input().split())) for _ in range(n)]
virus = [ [] for _ in range(k+1) ]
s,x,y = map(int,input().split())

for r in range(n):
    for c in range(n):
        if board[r][c] != 0:
            virus[board[r][c]].append((r,c))

mr = [1,-1,0,0]
mc = [0,0,1,-1]
for _ in range(s):
    for i in range(1,k+1):
        if not virus[i]:
            continue
        else :
            next = []
            for nr,nc in virus[i]:
                for l in range(4):
                    next_r = nr + mr[l]
                    next_c = nc + mc[l]
                    if 0 <= next_r < n and 0 <= next_c < n and board[next_r][next_c] == 0:
                        board[next_r][next_c] = i
                        next.append((next_r,next_c))
            virus[i] = next
        if board[x-1][y-1] != 0:
            break

print(board[x-1][y-1])

# continue와 break 사용시 주의하기 
```

```python
import sys
from collections import deque
input = sys.stdin.readline

def rotate(st,edn,edm,r):
    l = (edn - st) * 2 + (edm - st) * 2
    tmp = []
    
    # 맨 겉에 있는 줄부터 시계 반대 방향으로 하나씩 배열에 옮기기
    for i in range(edn - st):
        tmp.append(bd[st + i][st])

    for i in range(edm-st):
        tmp.append(bd[edn][st + i])

    for i in range(edn - st):
        tmp.append(bd[edn-i][edm])

    for i in range(edm - st):
        tmp.append(bd[st][edm - i])

    ans = deque([[] for _ in range(l)])
    mv = r % l

    # 돌리기
    for i in range(l):
        ans[(i + mv) % l] = tmp[i]

    # 돌아간 행렬 답안에 시계 반대 방행으로 붙여넣기
    for i in range(edn - st):
        answer[st + i][st] = ans.popleft()

    for i in range(edm-st):
        answer[edn][st + i] = ans.popleft()

    for i in range(edn - st):
        answer[edn - i][edm] = ans.popleft()

    for i in range(edm - st):
        answer[st][edm - i] = ans.popleft()
    return

n,m,r = map(int,input().split())
bd = [list(map(int,input().split())) for _ in range(n)]
answer = [[[]for _ in range(m)] for _ in range(n)]

# 가장 바깥에 있는 행렬부터 차례로 돌리기 
st, edn, edm = 0, n-1, m-1
for i in range(min(n,m) // 2):
    rotate(st+i, edn-i,edm-i,r)

for ans in answer:
    for a in ans:
        print(a,end= ' ')
    print()

```

```python
# 가로, 세로
a,b = map(int,input().split())
n,m = map(int,input().split())
board = [ [ 0 for _ in range(a)] for _ in range(b)]
robots = [[0,0,0]]
cmds = []
dirs = ['N','E','S','W']
mr = [-1,0,1,0]
mc = [0,1,0,-1]

for i in range(1,n + 1):
    c,r,dir = input().split()
    r = b - int(r)
    c = int(c) - 1
    dir = dirs.index(dir)
    robots.append([r,c,dir])
    board[r][c] = i

for _ in range(m):
    num, cmd, r = input().split()
    cmds.append([int(num),cmd,int(r)])


flag = True
for cd in cmds:

    num,cmd,rr = cd
    for _ in range(rr):
        r, c, dir = robots[num]
        if cmd == 'L':
            robots[num][2] = (dir - 1) % 4
        elif cmd == 'R':
            robots[num][2] = (dir + 1) % 4
        elif cmd == 'F':
            nr,nc = r + mr[dir], c + mc[dir]
            if not(0 <= nr < b and 0 <= nc < a):
                print('Robot %d crashes into the wall'%(num))
                flag = False
                break
            elif board[nr][nc] != 0:
                print('Robot %d crashes into robot %d'%(num,board[nr][nc]))
                flag = False
                break
            else:
                board[r][c] = 0
                board[nr][nc] = num
                robots[num] = [nr,nc,dir]
    if not flag:
        break


print(robots)
if flag: print('OK')

# 행렬 좌표 방향을 표현하는 값 표시 주의하기
```

```python
# 15980번 : 명상 방해꾼
sys.stdin = open('./input.txt')
input = sys.stdin.readline

n,m = map(int,input().split())
board = []
for _ in range(n):
    dir, times = input().split()
    board.append(list(map(lambda x : -int(x),times)) if dir == 'L' else list(map(int,times)))
sum_list = [sum([board[row][col] for row in range(n)]) for col in range(m)]
ans_list = [list(sum_list[col] - board[row][col] for col in range(m)) for row in range(n)]
for row in range(n):
    for col in range(m):
        if col == 0:
            continue
        else:
            ans_list[row][col] += ans_list[row][col-1]
answer = [max(list(map(abs,ans_list[row]))) for row in range(n)]
print(answer.index(min(answer)) + 1)
print(min(answer))
```
