```python
# 11279번 : 최대 힙 (직접구현)
heap = [-1]  
# add & heapify
def heap_add(num):
    heap.append(num)
    idx = len(heap) - 1
    # 넣은 원소가 제자리를 찾아가는 과정 
    while idx // 2 != 0:
        if abs(heap[idx]) < abs(heap[idx//2]) : 
            heap[idx], heap[idx//2] = heap[idx//2], heap[idx]
            idx = idx//2
        elif abs(heap[idx]) == abs(heap[idx//2]):
            if heap[idx] < heap[idx//2]:
                heap[idx], heap[idx//2] = heap[idx//2], heap[idx]
            idx = idx//2       
        else : break
                            
# delete & heapify
def heap_delete_max():
    if len(heap) == 1 : print(0)
    else :
        # heap 최대 값과 가장 뒤에 있는 값을 바꿈 
        heap[1], heap[-1] = heap[-1], heap[1]
        print(heap.pop())
        idx = 2
        m = len(heap) - 1
        # 바꾼 값이 heap에서 제자리를 찾아가는 과정 
        while idx <= len(heap) - 1:
            if idx < len(heap) - 1:
                if abs(heap[idx]) > abs(heap[idx + 1]):
                    idx += 1
                elif abs(heap[idx]) == abs(heap[idx + 1]):
                    if heap[idx] > heap[idx+1]:
                        idx += 1
            if abs(heap[idx//2]) < abs(heap[idx]) : break
            elif abs(heap[idx//2]) == abs(heap[idx]) : 
                if heap[idx//2] <= heap[idx] : break
            else : heap[idx], heap[idx//2] = heap[idx//2], heap[idx]
            idx *= 2

for _ in range(int(input())):
    num = int(input())
    if num:
        heap_add(num)
    else:
        heap_delete_max()
    print(heap)
```
